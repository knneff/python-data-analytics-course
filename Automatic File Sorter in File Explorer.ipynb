```python
import os
import shutil
```

```python
path = r"C:\Users\knmnl\Downloads"

files = os.listdir(path)

folders = {
    'csv files' : ['.csv'],
    'image files' : ['.png', '.jpeg', '.jpg', '.img'],
    'text files' : ['.txt']
}
```

```python
for folder in folders:
    folder_path = os.path.join(path, folder_name)
    os.makedirs(folder_path, exist_ok = True)
```

```python
for file in files:
    src_path = os.path.join(path, file)
    if os.path.isdir(src_path):
        continue
    _, ext = os.path.splitext(file.lower())
    for folder_name, folder_ext in folders.items():
        if ext in folder_ext:
            shutil.move(src_path, os.path.join(path, folder_name, file))
            break
        
```
